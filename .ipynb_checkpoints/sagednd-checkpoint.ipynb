```python
# Get Data
# Load a sage file that produces dnd ability roll data
load("dnd_ability_rolls.sage")

# Get the data we want to analyze
rolls = get_dnd_ability_rolls()
roll_sums = [dnd_ability_roll_sum(r) for r in rolls]
```

```python
# Elementary Analysis
# Compute some statistics
average_roll = mean(roll_sums)*1.0
roll_median = median(roll_sums)*1.0
roll_mode = mode(roll_sums)
roll_deviation = std(roll_sums, bias = False)*1.0
```

```python
# Print results of elementary analysis
average_roll; roll_mode; roll_median; round(roll_deviation);
print('\n')
print('average roll = '+str(average_roll)+'\n')
print('roll mode = '+str(roll_mode)+'\n')
print('roll median = '+str(roll_median)+'\n')
print('roll deviation = '+str(round(roll_deviation))+'\n')
```

```python
# Visualization Preparation
# Prepare a histogram
roll_hist = histogram(roll_sums, bins=16)
```

```python
#Display visualization
roll_hist
```

```python
# Looks like a bell curve
# Prepare a bell curve (by hand)
sigma = 3
mu = 12.3
my_bell(x) = len(roll_sums)*1/sigma/sqrt(2*pi) * exp(-(x-mu)^2/2/sigma^2)
approximate_bell = plot(my_bell(x), (x,3,22))
```

```python
# Display both the histogram and bell curve
sum([roll_hist, approximate_bell])
```
